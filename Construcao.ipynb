```python
# plotando a equação da reta

fig, ax = plt.subplots(figsize=(16,9))

ax.scatter(iris["petal width (cm)"], iris["petal length (cm)"], c=iris.target, cmap="PiYG")

ax.xaxis.set_label_text("Petal Width (cm)")
ax.yaxis.set_label_text("Petal Length (cm)")

x_reta1 = np.arange(0,3)
y_reta1 = -(10/3)*x_reta1 + 5

ax.plot(x_reta1, y_reta1, c="red")

ax.scatter(0.7, 3.5, c="r")


plt.show()
```

```python
# criando uma função classificadora com base na reta escolhida

def classificador(x, y):
    ruler = -(10/3)*x + 5
    
    if y < ruler:
        return 0
    elif y > ruler:
        return 1
    else:
        return -1
```

```python
# Importando o perceptron e fazendo o fit
from sklearn.linear_model import Perceptron

clf = Perceptron(tol=1e-3, random_state=0)
```

```python
# Selecionando o X e o y

X = iris[["petal width (cm)", "petal length (cm)"]]
y = iris.target
```

```python
# Fazendo o fit com o modelo
clf.fit(X, y)
```

```python
clf.score(X, y)
```

```python
# w1 e w2
clf.coef_
```

```python
# w0
clf.intercept_
```

```python
w0 + w1*x1 + w2*x2 = limite

para limite = 0, x1=x e x2=y

w2*y = - (w0 + w1*x)

y = -(w0*w1*x)/w2
```

```python
# Colocando tanto o perceptron quanto a reta que traçamos no gráfico
fig, ax = plt.subplots(figsize=(16,9))

ax.scatter(iris["petal width (cm)"], iris["petal length (cm)"], c=iris.target, cmap="PiYG")

ax.xaxis.set_label_text("Petal Width (cm)")
ax.yaxis.set_label_text("Petal Length (cm)")

x_reta1 = np.arange(0,3)
y_reta1 = -(10/3)*x_reta1 + 5

ax.plot(x_reta1, y_reta1, c="red", label="Olhômetro")

y_p = (-clf.intercept_[0] - clf.coef_[0][0]*x_reta1)/clf.coef_[0][1]

ax.plot(x_reta1, y_p, c="b", label="Perceptron")

ax.legend()

plt.show()
```

```python
# Adicionando uma nova base com os dados mostrados abaixo
base2 = pd.DataFrame([[0,0,3,0.75,0], [0,0,2.5,1,0]])
base2.columns = ['sepal length (cm)','sepal width (cm)','petal length (cm)','petal width (cm)','target']
```

```python
base2.head()
```

```python
# Adicionado esses dados ao dataset iris
iris2 = iris.copy()

iris2 = pd.concat([iris2, base2])
```

```python
iris2.tail()
```

```python
# Executando o Perceptron pra essa nova base
X2 = iris2[["petal length (cm)","petal width (cm)"]]
y2 = iris2.target

clf2 = Perceptron(tol=1e-3, random_state=0)
clf2.fit(X2, y2)
```

```python
fig, ax = plt.subplots()

scatter_plot = ax.scatter(iris2["petal length (cm)"],iris2["petal width (cm)"],c=iris2.target, cmap='PiYG')

ax.legend(handles=scatter_plot.legend_elements()[0], labels=scatter_plot.legend_elements()[1])

# Traçando novamente a reta gerada pelo perceptron
x_perc = np.arange(1,5)
y_perc = -((clf2.intercept_[0])/(clf2.coef_[0][1]))-((clf2.coef_[0][0])/(clf2.coef_[0][1]))*x_perc
ax.plot(x_perc,y_perc,c="b",linewidth=2.0)

# E a reta do nosso modelo
x3 = [1,3]
y3 = [2,0]
ax.plot(x3,y3,c="r",linewidth=2.0)

plt.show()
```
